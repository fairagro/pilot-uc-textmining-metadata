```python
import pandas as pd
```

```python
bonares = pd.read_csv("output_new.csv")
```

```python
filtered_bonares = bonares.dropna(ignore_index=True)
```

```python
filtered_bonares.head()
```

```python
filtered_bonares_year = bonares[(bonares['publication_year']  > 2018) & (bonares['publication_year'] < 2026) ]
```

```python
publication_count = len(filtered_bonares_year)
print(f"Number of publications after 2018: {publication_count}")
```

```python
from collections import Counter

# Sample keywords for crop and soil categories
crop_keywords_en = {'crop', 'plant', 'yield', 'photosynthesis', 'agriculture', 'harvest'}
soil_keywords_en = {'soil', 'nutrient', 'erosion', 'fertility', 'pH', 'compaction'}

crop_keywords_ger = {'ernte', 'anlage', 'ertrag', 'photosynthese', 'landwirtschaft', 'ernte'}
soil_keywords_ger = {'boden', 'nährstoff', 'erosion', 'fruchtbarkeit', 'pH-Wert', 'verdichtung'}

crop_keywords_fra = {'recadrer', 'usine', 'rendement', 'photosynthèse', 'agriculture', 'récolte'}
soil_keywords_fra = {'sol', 'nutritive','nutritif', 'érosion', 'fécondité', 'pH', 'compactage'}
```

```python
def classify_publication(keywords):
    keywords_set = set(keywords.lower().split())  

    crop_keywords = set(crop_keywords_en) | set(crop_keywords_ger) | set(crop_keywords_fra)
    soil_keywords = set(soil_keywords_en) | set(soil_keywords_ger) | set(soil_keywords_fra)

    is_crop = any(word in crop_keywords for word in keywords_set)
    is_soil = any(word in soil_keywords for word in keywords_set)

    if is_crop and is_soil:
        return 'Both'
    elif is_crop:
        return 'Crop'
    elif is_soil:
        return 'Soil'
    else:
        return 'Other'
```

```python
filtered_bonares_year_category = filtered_bonares_year.copy()
filtered_bonares_year_category['Category'] = filtered_bonares_year_category['keywords'].apply(classify_publication) 
```

```python
category_counts = filtered_bonares_year_category['Category'].value_counts()
```

```python
print("Publications by category:")
print(category_counts)
```
